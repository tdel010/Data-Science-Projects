```python
import pandas as pd #import the pandas 'toolbox'
```

```python
data = pd.read_csv('Realestate.csv') #read in the data file
```

```python
data.head() #display the first bit of the data to see what it contains
```

```python
data["houseAge"] = (data["houseAge"] - data["houseAge"].min()) / (data["houseAge"].max() - data["houseAge"].min()) #normalizes the houseAge data using the formula to convert values into a 0-1 scale
```

```python
data.head()
```

```python
data
```

```python
data.to_csv("Realestate_transformed.csv", index=False) #created a new csv file for the updated data and saved the first change to it
```

```python
data["houseAgeStandardized"] = data["houseAge"] #made a new column for the standardized houseAge data
```

```python
data.head()
```

```python
data.drop("numberOfConvenienceStores", axis=1, inplace=True) #removes the numberOfConvenienceStores column
```
